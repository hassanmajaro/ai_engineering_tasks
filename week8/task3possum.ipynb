# Statistical Analysis

```python
# putting cols into category
numerical_cols = [col for col in df2.select_dtypes(include=['int64', 'float64'])]
categorical_cols = [col for col in df2.select_dtypes(include=['object', 'category'])]
```

```python
for col in numerical_cols:
    min_value = df2[col].min()
    max_value = df2[col].max()
    mean = df2[col].mean()
    median = df2[col].median()
    mode = df2[col].mode().iloc[0]
    range = max_value - min_value

    print(f"For {col}")
    print(f"    Minimum value = {min_value}")
    print(f"    Maximum value = {max_value}")
    print(f"    Mean = {mean:.2f}")
    print(f"    Median = {median:.2f}")
    print(f"    Mode = {mode:.2f}")
    print(f"    Range = {range:.2f}\n")
```
