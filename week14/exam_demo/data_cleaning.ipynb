##### Task 1.4 Perform data cleaning to ensure the dataset is ready for analysis

```python
def clean_data(data):
    missing_data = data.isna().sum()
    if missing_data.sum() == 0:
        print("There are no missing data")
    else:
        data = data.fillna(0)
        print(f"Numbers of missing values before cleaning: {missing_data}")
        print(f"\nNumber of missing values after cleaning: {data.isna().sum()}")

    duplicates = data.duplicated().sum()
    if duplicates.sum() == 0:
        print("There are no duplicated data")
    else:
        data = data.drop_duplicates()
        print(f"Number of duplicated data before cleaning: {duplicates}")
        print(f"\nNumber of duplicated data after cleaning: {data.duplicated().sum()}")

clean_data(df)
```
